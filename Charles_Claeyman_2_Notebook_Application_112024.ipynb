```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import joblib
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

model = joblib.load("/content/drive/MyDrive/Projet12/modele_regression_logistique.pkl")
scaler = joblib.load("/content/drive/MyDrive/Projet12/scaler.pkl")

nouveaux_billets = pd.read_csv("/content/drive/MyDrive/Projet12/billets_production.csv")
# Si erreur de chargement, pensé à ajouter : sep=";"

colonnes = ['diagonal', 'height_left', 'height_right', 'margin_low', 'margin_up', 'length']
X_nouveau = nouveaux_billets[colonnes]

# Standardiser avec le scaler entraîné
X_nouveau_scaled = scaler.transform(X_nouveau)

# Prédire avec le modèle
predictions = model.predict(X_nouveau_scaled)

# Ajouter les prédictions au DataFrame
nouveaux_billets["prediction"] = predictions
nouveaux_billets["prediction"] = nouveaux_billets["prediction"].map({True: "Vrai billet", False: "Faux billet"})

# Sauvegarder les prédictions dans un fichier
nouveaux_billets.to_csv("/content/drive/MyDrive/Projet12/predictions_resultats.csv", index=False)
```
